```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes_binary_health_indicators.csv')
```

```python
df.shape
```

```python
# Checking duplicate rows in dataset df1 to eliminate them

duplicates = df[df.duplicated()]
print("Duplicate Rows : ",len(duplicates))
duplicates.head()
```

```python
# Eliminating 24206 duplicate rows from the dataset df
df.drop_duplicates(inplace= True)
```

```python
df.shape
```

```python
correlation_matrix = df[['Diabetes_binary', 'PhysActivity', 'Smoker', 'HvyAlcoholConsump']].corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Create a heatmap
plt.figure(figsize=(8, 6))  # Sets the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')  # Adds a title to the heatmap
plt.show()
```

```python
# Create a pair plot to visualize the relationships among diabetes, physical activity, smoking, and heavy alcohol consumption
sns.pairplot(df[['Diabetes_binary', 'PhysActivity', 'Smoker', 'HvyAlcoholConsump']], hue='Diabetes_binary', markers=["o", "s"])
plt.show()
```

```python
# Additionally, let's create individual scatter plots for each behavior with respect to diabetes
# Plot for Physical Activity vs. Diabetes
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='PhysActivity', y='Diabetes_binary', alpha=0.6)
plt.title('Physical Activity vs. Diabetes')
plt.xlabel('Physical Activity (1 = Active, 0 = Not Active)')
plt.ylabel('Diabetes (1 = Present, 0 = Absent)')
plt.show()
```

```python
# Plot for Smoking vs. Diabetes
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Smoker', y='Diabetes_binary', alpha=0.6)
plt.title('Smoking vs. Diabetes')
plt.xlabel('Smoker (1 = Smoker, 0 = Non-Smoker)')
plt.ylabel('Diabetes (1 = Present, 0 = Absent)')
plt.show()
```

```python
# Plot for Heavy Alcohol Consumption vs. Diabetes
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='HvyAlcoholConsump', y='Diabetes_binary', alpha=0.6)
plt.title('Heavy Alcohol Consumption vs. Diabetes')
plt.xlabel('Heavy Alcohol Consumption (1 = Yes, 0 = No)')
plt.ylabel('Diabetes (1 = Present, 0 = Absent)')
plt.show()
```

```python
#Checking The relation B/W Smoker and Diabetes

pd.crosstab(df.Smoker,df.Diabetes_binary).plot(kind="bar",figsize=(5,4))

plt.title('Diabetes Disease Frequency for Smoker')
plt.xlabel("Smoker")
plt.ylabel('Frequency')
plt.show()
```

```python
#Checking The relation B/W HvyAlcoholConsump and Diabetes

pd.crosstab(df.HvyAlcoholConsump,df.Diabetes_binary).plot(kind="bar",figsize=(5,4))

plt.title('Diabetes Disease Frequency for HvyAlcoholConsump')
plt.xlabel("HvyAlcoholConsump")
plt.ylabel('Frequency')
plt.show()
```

```python
sns.catplot(x="Smoker" , y ="HvyAlcoholConsump" , data = df , hue="Diabetes_binary"  , kind="bar"  );
plt.title("Relation b/w Smoker ,HvyAlcoholConsump and Diabetes")
```
