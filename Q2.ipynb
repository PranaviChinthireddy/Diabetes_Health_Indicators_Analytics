```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset to explore and understand its structure
data_path = '/Users/csuftitan/Desktop/diabetes_binary_health_indicators.csv'
data = pd.read_csv(data_path)
data = data.drop_duplicates()


# Display the first few rows of the dataset and information about the columns
data.head(), data.info()
```

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()
correlation_matrix['Diabetes_binary']

# Sort the correlation values with Diabetes_binary in descending order
sorted_correlations = correlation_matrix['Diabetes_binary'].sort_values(ascending=False)
sorted_correlations
```

```python
# Extracting the correlation coefficients with 'Diabetes_binary'
diabetes_correlations = correlation_matrix['Diabetes_binary'].drop('Diabetes_binary')  # excluding self-correlation

# Plotting a horizontal bar chart for the correlation coefficients using the 'coolwarm' palette
plt.figure(figsize=(10, 6))
colors = sns.color_palette("coolwarm", len(diabetes_correlations))
diabetes_correlations.sort_values().plot(kind='barh', color=colors)
plt.title('Correlation Coefficients with Diabetes Status')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Health Indicators')
plt.axvline(0, color='black', linewidth=0.8)  # Adding a line at zero for reference
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plotting the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Diabetes and Health Indicators')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
sns.set(style="whitegrid")

# Prepare multiple figures
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))  # Adjusted for fewer plots

# High Blood Pressure vs Diabetes
sns.countplot(x='HighBP', hue='Diabetes_binary', data=data, ax=axes[0, 0])
axes[0, 0].set_title('High Blood Pressure and Diabetes')
axes[0, 0].set_xlabel('High Blood Pressure')
axes[0, 0].set_ylabel('Count')

# High Cholesterol vs Diabetes
sns.countplot(x='HighChol', hue='Diabetes_binary', data=data, ax=axes[0, 1])
axes[0, 1].set_title('High Cholesterol and Diabetes')
axes[0, 1].set_xlabel('High Cholesterol')
axes[0, 1].set_ylabel('Count')

# Difficulty Walking vs Diabetes
sns.countplot(x='DiffWalk', hue='Diabetes_binary', data=data, ax=axes[1, 0])
axes[1, 0].set_title('Difficulty Walking and Diabetes')
axes[1, 0].set_xlabel('Difficulty Walking')
axes[1, 0].set_ylabel('Count')

# Heart Disease or Attack vs Diabetes
sns.countplot(x='HeartDiseaseorAttack', hue='Diabetes_binary', data=data, ax=axes[1, 1])
axes[1, 1].set_title('Heart Disease/Attack and Diabetes')
axes[1, 1].set_xlabel('Heart Disease or Attack')
axes[1, 1].set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'filtered_data' is your DataFrame
fruit_diabetes_crosstab = pd.crosstab(data['Fruits'], data['Diabetes_binary'])

# Set up the figure for pie charts
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Labels and colors for the pie charts
labels = ['No Diabetes', 'Diabetes']
colors = ['#4D77FF', '#FF6F61']

# Pie chart for individuals who do not consume fruits regularly
no_fruits_data = fruit_diabetes_crosstab.iloc[0]
axes[0].pie(no_fruits_data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axes[0].set_title('Diabetes Status for No Fruit Consumption')

# Pie chart for individuals who consume fruits regularly
fruits_data = fruit_diabetes_crosstab.iloc[1]
axes[1].pie(fruits_data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axes[1].set_title('Diabetes Status for Regular Fruit Consumption')

# Adjust layout and show plot
plt.tight_layout()
plt.show()

# Assuming 'filtered_data' is your DataFrame
veggie_diabetes_crosstab = pd.crosstab(data['Veggies'], data['Diabetes_binary'])

# Set up the figure for pie charts
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Labels and colors for the pie charts
labels = ['No Diabetes', 'Diabetes']
colors = ['#4D77FF', '#FF6F61']

# Pie chart for individuals who do not consume vegetables regularly
no_veggies_data = veggie_diabetes_crosstab.iloc[0]
axes[0].pie(no_veggies_data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axes[0].set_title('Diabetes Status for No Vegetable Consumption')

# Pie chart for individuals who consume vegetables regularly
veggies_data = veggie_diabetes_crosstab.iloc[1]
axes[1].pie(veggies_data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axes[1].set_title('Diabetes Status for Regular Vegetable Consumption')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load your data
data

# Filtering data for plotting
data = data[['Diabetes_binary', 'BMI']]

# Plotting
plt.figure(figsize=(12, 8))
sns.histplot(data = data, x='BMI', hue='Diabetes_binary', kde=True, element="step")
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.legend(title='Diabetes Status', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

```python
# Histograms of BMI split by diabetes status
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="BMI", hue="Diabetes_binary", element="step", stat="density", common_norm=False, palette="viridis")
plt.title('Distribution of BMI by Diabetes Status')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Density')
plt.legend(title='Diabetes Status', labels=['No Diabetes', 'Diabetes'])
plt.show()
```
